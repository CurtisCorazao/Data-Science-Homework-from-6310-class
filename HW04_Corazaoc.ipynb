Section 6b: Create a PCA for df_comedy_pivot
OBJECTIVE: Time to reduce the number of dimensions from 243,237 (number of columns in
df_comedy_pivot) to a more reasonable number that still captures the explained variance.
• Create a PCA with n_pcs equal to 85.
o Name your PCA for df_comedy_pivot as df_comedy_pca.
• Add headings for each column in df_comedy_pca. (Should be PC1, PC2, etc.)
• Create a DataFrame that shows the explained_ratio and explained_ratio_sum (name
of DataFrame can be explained_ratio like the videos in the Scenario for PCA).
o Why n_pcs 85? After looking at the explained variance ratio, n_pcs = 85
explains 70% of the variance, which is a general rule (by Dr. A) to use so that
you do not lose too much of the variance explained.

```python
# CodeBlock SubSection 6b
# creating pca
from sklearn.decomposition import PCA
# Set number of principal components for model
n_pcs = 85
model = PCA(n_components=n_pcs)
# Fit the model - return projected data
df_comedy_pca = model.fit_transform(df_comedies_pivot)
#Create a DataFrame of the projected data
df_comedy_pca = pd.DataFrame(df_comedy_pca)
df_comedy_pca.head()
```

```python
# CodeBlock SubSection 6b
# giving column headings
names = []
for x in range(n_pcs):
    names.append('PC%d'%(x+1))
    
df_comedy_pca.columns = names
df_comedy_pca.head()
```

```python
# CodeBlock SubSection 6b
# Create a DataFrame that shows the explained_ratio and explained_ratio_sum
explained_ratio = model.explained_variance_ratio_
explained_ratio = pd.DataFrame(explained_ratio).reset_index()
explained_ratio = explained_ratio.rename(columns = {0: 'explained_ratio', 'index': 'PC'})
explained_ratio['explained_ratio_sum']= np.cumsum(explained_ratio.explained_ratio)
explained_ratio
```

Section 7a: Create a correlation matrix from SVD
OBJECTIVE: For some reason, I just do not believe the results from the PCA. Why would this
be the case? If you go back to the Scenarios/Presentations, you will see that one method for
Recommender Systems does a better job than the other when the data is sparse.
(Sparse refers to data where most of the values are NaNs. Since most users only rate a few
movies, most of the original data in corr_comedy_pivot (prior to filling in NaNs with 0) were
NaN. Therefore, we should look at SVD as a better option for a recommender system.)
• Create a SVD with n_components equal to 85.
o Name your SVD for df_comedies_pivot as resultant_matrix_SVD.
• Create a correlation matrix using resultant_matrix_SVD. Name it
corr_mat_SVD_comedy.
o Change corr_mat_SVD_comedy into a DataFrame with the columns =
comedy_names.
o Set the corr_mat_SVD_comedy index = comedy_names.
o Should be similar to Fgiure 10.

```python
# CodeBlock SubSection 7a
# importing libraries and creating resultant_matrix_SVD

from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=85)
resultant_matrix_SVD = SVD.fit_transform(df_comedies_pivot)
resultant_matrix_SVD.shape
```

```python
# CodeBlock SubSection 7a
# adding column names to matrix for corr_mat_SVD_comedy
corr_mat_SVD_comedy = np.corrcoef(resultant_matrix_SVD)
corr_mat_SVD_comedy = pd.DataFrame(corr_mat_SVD_comedy, columns=(comedy_names))
corr_mat_SVD_comedy.index = comedy_names
corr_mat_SVD_comedy.head()
```
